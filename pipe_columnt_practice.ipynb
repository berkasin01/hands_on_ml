```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

num_cols = ["Age", "Sleep Duration", "Physical Activity Level", "Stress Level", "Heart Rate", "Daily Steps"]
cat_cols = ["Gender", "Occupation", "BMI Category"]
```

```python
num_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy="median")),
    ("stan_scaler",StandardScaler()),
])
```

```python
cat_pipeline = Pipeline([
("imputer",SimpleImputer(strategy="most_frequent")),
("ohe",OneHotEncoder(handle_unknown="ignore")),
])
```

```python
column_transformer = ColumnTransformer([
    ("numbers",num_pipeline,num_cols),
    ("texts",cat_pipeline,cat_cols),
])
```

```python
column_transformer.fit_transform(df)
```
