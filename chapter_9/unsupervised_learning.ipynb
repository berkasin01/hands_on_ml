```python
from sklearn.datasets import make_blobs
import numpy as np

# Clustering dataset
X_blobs, y_blobs = make_blobs(n_samples=500, centers=5, cluster_std=0.8, random_state=42)

# Moons dataset (nonlinear structure)
from sklearn.datasets import make_moons
X_moons, y_moons = make_moons(n_samples=500, noise=0.05, random_state=42)

# Anisotropic blobs (varied density)
X_aniso, y_aniso = make_blobs(n_samples=500, centers=3, random_state=42)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X_aniso, transformation)
```

```python
from sklearn.cluster import KMeans

k=5 
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X_aniso)
```

```python
y_pred
```

```python
kmeans.cluster_centers_
```

```python
X_new=np.array([[0,2],[3,2],[-3,3],[-3,2.5]])
kmeans.predict(X_new)
kmeans.transform(X_new)
```

```python
from sklearn.cluster import MiniBatchKMeans

minibatch_kmeans = MiniBatchKMeans(n_clusters=5)
minibatch_kmeans.fit(X_aniso)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_blobs)
    scores.append(silhouette_score(X_blobs, labels))

import matplotlib.pyplot as plt
plt.plot(k_range, scores, marker='o')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs k')
plt.show()
```

```python
from sklearn.datasets import load_digits 

X_digits, y_digits = load_digits(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_digits,y_digits)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
log_reg.score(X_test,y_test)
```

```python
X_digits
```

```python
y_digits
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(
[
    ("kmeans",KMeans(n_clusters=50)),
    ("log_reg",LogisticRegression()),
]   
)
```

```python
pipeline.fit(X_train,y_train)
```

```python
pipeline.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(2,100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
grid_clf.fit(X_train,y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test,y_test)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X,y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
dbscan.core_sample_indices_
```

```python
dbscan.components_
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.5,0],[0,0.05],[1,-0.1],[2,1]])
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2] = -1
y_pred.ravel()
```
