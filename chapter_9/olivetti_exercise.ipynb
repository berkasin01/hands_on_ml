```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
olivetti_faces = fetch_olivetti_faces()
```

```python
olivetti_faces.images.shape
```

```python
olivetti_faces.target.shape
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

X = olivetti_faces.images.reshape(len(olivetti_faces.images), -1)
print(X.shape) 

r = range(2,11)

inertias = []
for i in r:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertia = kmeans.inertia_
    inertias.append(inertia)

plt.plot(r,inertias,marker='o')
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

r = range(2,11)

s_scores = []
for i in r:
    kmeans = KMeans(n_clusters=i)
    s_score = silhouette_score(X,kmeans.fit_predict(X))
    s_scores.append(s_score)

plt.plot(r,s_scores,marker='o')
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
best_k = KMeans(n_clusters=5)
best_k.fit(X)
clusters = best_k.labels_
clusters
```
