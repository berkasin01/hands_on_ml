```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
olivetti_faces = fetch_olivetti_faces()
```

```python
olivetti_faces.images.shape
```

```python
olivetti_faces.target.shape
```

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

X = olivetti_faces.images.reshape(len(olivetti_faces.images), -1)
print(X.shape) 

r = range(2,11)

inertias = []
for i in r:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertia = kmeans.inertia_
    inertias.append(inertia)

plt.plot(r,inertias,marker='o')
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

r = range(2,11)

s_scores = []
for i in r:
    kmeans = KMeans(n_clusters=i)
    s_score = silhouette_score(X,kmeans.fit_predict(X))
    s_scores.append(s_score)

plt.plot(r,s_scores,marker='o')
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
best_k = KMeans(n_clusters=5)
best_k.fit(X)
clusters = best_k.labels_
clusters
```

```python
olivetti_faces.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, olivetti_faces.target, test_size=0.2, random_state=42, stratify=olivetti_faces.target
)
```

```python
y_train
```

```python
from xgboost import XGBClassifier
```

```python
xgb_class = XGBClassifier()
xgb_class.fit(X_train,y_train)
```

```python
y_hat = xgb_class.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test,y_hat)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_hat))
```
